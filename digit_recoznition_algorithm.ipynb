```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('emnist-mnist-train.csv')
```

```python
data
```

```python
data.shape
```

```python
data_man = data.drop('4',axis = 1)
for i in range(len(data)):
    row = np.array(data.iloc[i,:].drop('4',axis = 0)).reshape(28,28).T
    row_c = row.reshape(1,784)
    data_man.iloc[i,:] = row_c
```

```python
data_man['label'] = data['4']
```

```python
data_man
```

```python
plt.matshow(np.array(data_man[data_man['label'] == 3].iloc[200,:].drop('label',axis = 0)).reshape(28,28))
```

```python
plt.matshow(np.array(data_man[data_man['label'] == 2].iloc[200,:].drop('label',axis = 0)).reshape(28,28))
```

```python
data_test = pd.read_csv('emnist-mnist-test.csv')
```

```python
data_test
```

```python
data_man_test = data_test.drop('1',axis = 1)
for i in range(len(data_test)):
    row = np.array(data_test.iloc[i,:].drop('1',axis = 0)).reshape(28,28).T
    row_c = row.reshape(1,784)
    data_man_test.iloc[i,:] = row_c
```

```python
data_man_test['label'] = data_test['1']
```

```python
data_man_test
```

```python
plt.matshow(np.array(data_man_test[data_man_test['label'] == 3].iloc[100,:].drop('label',axis = 0)).reshape(28,28))
```

```python
x_train = data_man.drop('label',axis = 1)
y_train = data_man['label']
```

```python
x_test = data_man_test.drop('label',axis = 1)
y_test = data_man_test['label']
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
x_train = ss.fit_transform(x_train)
x_test = ss.fit_transform(x_test)
```

```python
x_train = pd.DataFrame(x_train)
x_test = pd.DataFrame(x_test)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(activation = 'relu',alpha = 0.1,solver = 'adam',hidden_layer_sizes = (15,15,15))
```

```python
mlp.fit(x_train,y_train)
```

```python
mlp.score(x_test,y_test)
```

```python
mlp2 = MLPClassifier(activation = 'relu',alpha = 0.1,solver = 'adam',hidden_layer_sizes = (20,20,20))
```

```python
mlp2.fit(x_train,y_train)
```

```python
mlp2.score(x_test,y_test)
```

```python
mlp3 = MLPClassifier(activation = 'relu',alpha = 0.1,solver = 'adam',hidden_layer_sizes = (50,50,50))
```

```python
mlp3.fit(x_train,y_train)
```

```python
mlp3.score(x_test,y_test)
```

```python
mlp4 = MLPClassifier(activation = 'relu',alpha = 0.1,solver = 'adam',hidden_layer_sizes = (100,100,100))
```

```python
mlp4.fit(x_train,y_train)
```

```python
mlp4.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,mlp4.predict(x_test))
```

```python
import seaborn as sns
plt.figure(figsize = (20,20))
sns.heatmap(cm,annot = True)
```
